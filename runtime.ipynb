```python
import numpy as np
yunwei=np.array([[1.59,2.3,3.14,3.9,4.7]
                 ,[1.88,2.99,3.66,4.75,5.67],
                 [5.54,10.49,15.68,20.23,25.28]
])
sci=np.array([[2.97,4.99,11.35,13.21,14.9],[8.04,10.52,12.98,15.84,18.33],[5.4,8.78,15.67,18.81,22.54]])
ratio=yunwei/sci
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
color=['r','b','k']
label=[1e-2,6e-8,1.1e-16]
for i in range(len(yunwei)):
    ax.plot(yunwei[i],marker='o',color=color[i],label=label[i])
    ax.plot(sci[i],marker='v',color=color[i],label=label[i])
ax.legend()
```
